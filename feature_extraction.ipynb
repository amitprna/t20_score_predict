```python
city = np.where(df['city'].isnull(),df['venue'].apply(lambda x : x.split()[0]),df['city'])
```

```python
df['city'].isnull().sum() 
```

```python
# venue column is no longer required
df.drop(columns=['venue'],inplace=True)
```

```python
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df
```

```python
df['ball_thrown']=(df['ball'].apply(lambda x : str(x).split('.')[0]).astype(int)*6) + df['ball'].apply(lambda x : str(x).split('.')[1]).astype(int)
```

```python
df
```

```python
df['player_dismissed']=df['player_dismissed'].apply(lambda x : 1 if len(x)>1 else 0).astype(int)
```

```python
df['wicket_left'] = 10-df.groupby('match_id').cumsum()['player_dismissed']
```

```python
df
```

```python
df = df.groupby('match_id').sum()['runs'].reset_index().merge(df,on='match_id')
```

```python
df.drop(columns=['runs_y','player_dismissed','final_run'],inplace=True)
```

```python
df.drop(columns=['ball'],inplace=True)
```

```python
df.rename({'runs_x':'final_score'},axis=1,inplace=True)
```

```python
df
```

# Model Bulding

```python
from sklearn.model_selection import train_test_split

X = df.drop(['final_score'],axis=1)
y = df['final_score']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

from sklearn.metrics import r2_score,mean_absolute_error
```

```python
trf = ColumnTransformer([('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])],
                        remainder='passthrough')
```

## Random Forest

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',RandomForestRegressor(n_estimators=100,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=15))
]
)
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

## XGBoost

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler()),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
pickle.dump(pipe,open('pipe.pkl','wb'))
```
